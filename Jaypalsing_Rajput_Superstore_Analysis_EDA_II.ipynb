# 2. What does the customer rating look like and can you also comment on its skewness? 

```python
sns.distplot(supmdf['Rating'])
plt.axvline(x=np.mean(supmdf['Rating']), c='red', ls='-.', label='mean')
plt.axvline(x=np.percentile(supmdf['Rating'],25),c='green', ls='--', label = '25th percentile:Q1')
plt.axvline(x=np.percentile(supmdf['Rating'],75),c='blue', ls='--',label = '75th percentile:Q3' )
plt.legend()
plt.title("customer rating about Supermarket and Skewness")
```

The rating distribution looks symmetric and there is no skewness on the left or right side of the distribution. I plot the mean and the percentiles to show clear picture. The red dashed dotted line is the mean and the green and blue lines indicate the 25th and 75th percentiles respectively which is equali distributed.The mean rating is equals to 7 given by the customers across products.

#3.	What can you say about the aggregate sales across the branches?

# 4.	Can you see any time trend in gross income? 

```python
#To plot the time trend in gross income first i grouped dates as per gross income
fig, ax = plt.subplots(figsize=(15, 5))
sns.lineplot(x=supmdf.groupby(supmdf['Date']).mean().index,
             y=supmdf.groupby(supmdf['Date']).mean()['gross_income'])
plt.xticks(rotation=90)
plt.title("Time trend in gross income")
```

**Observations**
With help of lineplot and datewise gross income trend No particular time trend is observed except for 12 Feb 2019 was spotted higher and 13 Feb 2019 was spooted lower gross income days. Overall it remains at a certain average level.

# 5.	Which product line helps you generate the most income?

```python
plt.figure(figsize=(20,8))
sns.barplot(x='product_line', y='gross_income', data=catgdf)
plt.title("Higher income product line(Category)")
```

**Observations**
The Food and beverages,sports and travel	product line helps to generate higher gross income.Remaining product lines also performing well.

#8.	Which day of the week has maximum sales? 

```python
#converting Date column type from object to Date
import datetime as dt
supmdf['Date'] = pd.to_datetime(supmdf['Date'])
```

```python
supmdf.info()
```

```python
#creating new weekday column and store textual name of day
supmdf['weekday'] = supmdf['Date'].dt.day_name()
supmdf.head(3)
```

```python

plt.figure(figsize=(8, 6))
sns.barplot(x=supmdf.groupby(supmdf['weekday']).mean().index,
             y=supmdf.groupby(supmdf['weekday']).mean()['Total'])#Grouping days by total sales
plt.title("Maximum sales by day of the week ")
```

**Observations**
Maximum sales in A  week taken place on Saturday and second follwer acording to total sales was Sunday

#9.Which hour of the day is busiest? 

```python
#Column type conversion from Object to time
supmdf['Time'] = pd.to_datetime(supmdf['Time'])
```

```python
supmdf.info()
```

```python
#creating new Hour column 

supmdf['Hour'] = supmdf['Time'].dt.hour
supmdf.head(3)
```

```python
#Grouping Hour by total Sales and display in descending order
bhour=supmdf.groupby(supmdf['Hour']).mean()['Total'].sort_values(ascending=False)
bhour
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x=supmdf.groupby(supmdf['Hour']).mean().index,
             y=supmdf.groupby(supmdf['Hour']).mean()['Total'])#Grouping Hours by total sales
plt.title('Busiest Hour in Super Market')
```

**Observations**
Maximum sales in day  was taken place on time of 14:00pm and second follower acording to total sales as per time was 19:00pm
